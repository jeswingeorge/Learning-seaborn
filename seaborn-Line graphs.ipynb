```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

## Line graph

```python
# drawing a simple line graph
ax = sns.lineplot(data = iris, x = iris.index, y = "sepal_length")
```

```python
# drawing a simple line graph ralting to different groups using hue
ax = sns.lineplot(data = iris, x = iris.index, y = "sepal_length", hue = "species")
```

## How to fill area between 2 line graphs?

```python
linear_data = list(range(11))
exponential_data = [x**2 for x in linear_data]
```

```python
fig = plt.figure(figsize  = (10, 8))
ax = sns.lineplot(x = list(range(11)), y = linear_data, label = 'Linear Data')
ax = sns.lineplot(x = list(range(11)), y = exponential_data, label = 'Exponential Data')
plt.fill_between(x = list(range(11)), y1 = linear_data, y2 = exponential_data, color="skyblue", alpha=0.8)
plt.legend(fontsize  = 'large')
plt.show()
```
