```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
sns.set(rc = {'figure.figsize':(10, 8)}, font_scale=1.5)
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

## [Seaborn swarmplot](https://seaborn.pydata.org/generated/seaborn.swarmplot.html)

-  categorical scatter plot

Draw a categorical scatterplot with non-overlapping points.

```python
# Draw a single horizontal swarm plot:
sns.swarmplot(x = tips['total_bill']);
```

```python
# Group the swarms by a categorical variable:
sns.swarmplot(data = tips, x = 'sex', y = 'tip');
```

```python
# Draw horizontal swarms:
sns.swarmplot(data = tips, x = 'total_bill', y = 'day');
```

```python
# Color the points using a second categorical variable:
sns.swarmplot(data = tips, y = 'total_bill', x = 'day', hue = 'sex');
```

```python
# Split each level of the hue variable along the categorical axis using dodge=True:
sns.swarmplot(x="day", y="total_bill", hue="smoker", data=tips, dodge=True);
```

```python
# Control swarm order by passing an explicit order:
sns.swarmplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"]);
```

```python
# Plot using larger points:
sns.swarmplot(x="time", y="tip", data=tips, size=6);
```

```python
# Draw swarms of observations on top of a box plot:

# Color for all of the elements, or seed for a gradient palette. -> color = ".2"
sns.boxplot(x="tip", y="day", data=tips);
sns.swarmplot(x="tip", y="day", data=tips, color = ".2");
```

```python
# Draw swarms of observations on top of a violin plot:
ax =  sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
ax = sns.swarmplot(x="day", y="total_bill", data=tips, color="white", edgecolor="gray")
```

Use [catplot()](https://seaborn.pydata.org/generated/seaborn.catplot.html#seaborn.catplot) to combine a swarmplot() and a FacetGrid. This allows grouping within additional categorical variables. Using catplot() is safer than using [FacetGrid](https://seaborn.pydata.org/generated/seaborn.FacetGrid.html#seaborn.FacetGrid) directly, as it ensures synchronization of variable order across facets.

```python
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=tips, kind="swarm", height=6, aspect=.7);
```
