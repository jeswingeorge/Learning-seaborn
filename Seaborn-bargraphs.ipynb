```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
sns.set_context(rc = {'figure.figsize':(11, 9), "font.size":15, "axes.labelsize":20}, font_scale=2)
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

## [Seaborn bargraph](https://seaborn.pydata.org/generated/seaborn.barplot.html)

```python
# Draw a set of vertical bar plots grouped by a categorical variable:
sns.barplot(x ='day', y = 'total_bill', data = tips);
```

```python
# Draw a set of vertical bars with nested grouping by a two variables:
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex');
```

```python
# Draw a set of horizontal bars:
sns.barplot(x="tip", y="day", data=tips);
```

```python
# Control bar order by passing an explicit order:
sns.barplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"]);
```

```python
# Control bar order by passing an explicit order without ci
sns.barplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"], ci = None);
```

```python
# Use median as the estimate of central tendency:
sns.barplot(x="day", y="tip", data=tips, estimator=np.median);
```
